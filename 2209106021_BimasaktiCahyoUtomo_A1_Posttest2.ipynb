# 5. Tampilkan jumlah nilai dari semua attribute yang bernilai object (contoh : Jenis Kelamin, Agama).

```python
hitung = []

for column in df.columns:
    if df[column].dtype == "object":
        penghitung = df[column].value_counts().reset_index()
        penghitung.columns = ['Nilai', 'Jumlah']
        penghitung['Atribut'] = column
        hitung.append(penghitung)
        
jumlah_nilai = pd.concat(hitung, ignore_index=True)
neworder = ["Atribut","Nilai","Jumlah"]
jumlah_nilai.reindex(columns=neworder)
```

### Penjelasan
1. **hitung = []** : digunakan untuk menyimpan data hasil dari perulangan
2. **jumlah_nilai** : variabel yang mewakili dataframe baru
3. **neworder** : varibal digunakan untuk menyimpan nama dan urutan kolom agar lebih rapi
4. **jumlah_nilai.reindex(columns=neworder)** :untuk mengurutkan kolom sesuai variabel **neworder**

#### Proses :

1. Iterasi melalui kolom kolom DataFrame **df**:
     **for column in df.columns:**: Ini adalah loop yang iterasi melalui semua nama kolom dalam DataFrame **df**. <br>

2. Memeriksa jenis data kolom:
     **if df[column].dtype == "object":** : Kode ini memeriksa apakah jenis data dari kolom yang sedang diperiksa adalah objek (yaitu, berjenis data string atau kategori). Jika ya, maka kode di dalamnya akan dijalankan. <br>

3. Menghitung jumlah nilai unik:
     **penghitung = df[column].value_counts().reset_index()**: Kode ini menghitung jumlah kemunculan (value counts) dari nilai nilai unik dalam kolom yang berjenis data objek dan menghasilkan sebuah DataFrame yang berisi dua kolom: 'index' (nilai unik) dan 'column' (jumlah kemunculan). <br>

4. Mengubah nama kolom:
     **penghitung.columns = ['Nilai', 'Jumlah']**: Ini mengubah nama kolom dalam DataFrame **penghitung** sehingga 'index' menjadi 'Nilai' dan 'column' menjadi 'Jumlah'. <br>

5. Menambahkan kolom 'Atribut':
     **penghitung['Atribut'] = column**: Ini menambahkan sebuah kolom 'Atribut' ke DataFrame **penghitung**, yang berisi nama kolom yang sedang diperiksa. <br>

6. Menyimpan DataFrame dalam list:
     **hitung.append(penghitung)**: DataFrame **penghitung** yang telah dihasilkan untuk kolom tertentu akan dimasukkan ke dalam list **hitung**. <br>

7. Menggabungkan hasil perhitungan:
     **jumlah_nilai = pd.concat(hitung, ignore_index=True)**: Setelah iterasi selesai, kode ini menggabungkan semua DataFrames yang ada dalam list **hitung** menjadi satu DataFrame tunggal, yaitu **jumlah_nilai**. Opsi **ignore_index=True** digunakan untuk mengatur ulang indeks DataFrame hasil gabungan sehingga indeks menjadi berurutan.<br>

8. Mengganti urutan kolom:
     **neworder = ["Atribut","Nilai","Jumlah"]**: Ini adalah inisialisasi urutan kolom yang baru. <br>

9. **jumlah_nilai.reindex(columns=neworder)**: Baris ini mengganti urutan kolom dalam DataFrame **jumlah_nilai** sesuai dengan urutan yang ditentukan dalam list **neworder**. <br>

Hasilnya adalah DataFrame **jumlah_nilai** yang berisi tiga kolom: 'Atribut' (nama kolom asli), 'Nilai' (nilai unik dari kolom), dan 'Jumlah' (jumlah kemunculan nilai unik dalam kolom tersebut). DataFrame ini merangkum statistik tentang nilai nilai dalam kolom kolom objek dalam DataFrame asli **df**.
